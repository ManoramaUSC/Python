Given a two integer numbers return their product and if the product isgreater than 1000, then return their sum
Given: number1 = 20number2 = 30 Expected Output: The result is 600 
Given: number1 = 40 number2 = 30 Expected Output: The result is 70

```python
def fun(number1, number2): 
    mult = number1 * number2 
    if mult > 1000: 
        return number1 + number2 
    else: 
         return mult 
 
ans = fun(20, 30) 
print("The answer is->", ans) 
 
ans = fun(40, 30) 
print("The answer is->", ans) 
```
