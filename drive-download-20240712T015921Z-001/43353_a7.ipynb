```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df.head()
```

```python
df.isnull()
```

```python
df['target'] = iris.target
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'),df.target,test_size=0.2)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

#### Remark: I got accuracy of 96% using 100 trees

#### Q. Build decision tree model to predict survival of titanic based on certain parameters

In this file using following columns build a model to predict if person would survive or not,

1. Pclass 2. Sex 3. Age 4. Fare

Calculate score of your model

```python
import pandas as pd
import math
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('C:/Users/pooja/Downloads/titanic.csv')
df.head()
```

```python
df = df[['Survived','Pclass','Sex','Age','Fare']]
df.head()
```

```python
df.isnull().sum()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df['Survived']
df.head(10)
```

```python
inputs['Age'] = inputs['Age'].fillna(math.floor(inputs['Age'].mean()))
inputs.head(10)
```

```python
le = LabelEncoder()
inputs['Sex'] = le.fit_transform(inputs['Sex'])
inputs.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
